```python
from stingray import lightcurve
import numpy as np
from stingray.bispectrum import Bispectrum
from scipy.io import loadmat, savemat 
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
# 1. Read a time stacked image.. with vertical axis as distance and horizontal axis of time - currently done manually 
matstack = loadmat('bispectral.mat')

#bispectral.mat','grom_all_1', 'grom_all_2','spd_all_1','spd_all_2',........
#                      'Su_post_all','Su_neg_all');
#                   'f_MVCO','f_neg_MVCO',....);
#                      'f_post_MVCO');
                  
xx = matstack['f_MVCO']
times=1/xx 

dt = 0.0001  # seconds
dt=0.015625  ;

freq = 1 #Hz
exposure = 1  # seconds
times = np.arange(0, exposure, dt)  # seconds


counts= matstack['f_neg_MVCO']
print(times)
counts=np.squeeze(counts)
print(np.size(times))

#print(np.type(counts))
#Sv  = matstack['Sv_subset']
```

```python

#times = np.arange(1,11)
#counts = np.array([2, 1, 3, 4, 2, 5, 1, 0, 2, 3])


lc = lightcurve.Lightcurve(times,counts)
print(lc)
lc.plot(labels=['times','counts'])

#lc.counts
```

```python
bs = Bispectrum(lc)
```

```python
bs.freq
```

```python
p = bs.plot_mag()
p.show()
```

```python
p = bs.plot_cum3()
# negative is interval 1 , more diffused freq-freq plot 
```
